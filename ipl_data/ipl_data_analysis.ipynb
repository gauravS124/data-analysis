Which team has won/lost maximum number of matches while chasing?

```python
played_last=[]
losers=[]
for i in range(len(matches)):
    to_be_appended=matches.iloc[i]["team2"] if matches.iloc[i]["winner"]==matches.iloc[i]["team1"] else matches.iloc[i]["team1"]
    losers.append(to_be_appended)

    if(matches.iloc[i]["toss_decision"]=="field"):
        played_last.append(matches.iloc[i]["toss_winner"])
    else:
        to_be_appended=matches.iloc[i]["team2"] if matches.iloc[i]["team1"]==matches.iloc[i]["toss_winner"] else matches.iloc[i]["team1"]
        played_last.append(to_be_appended)
        
matches["chasing"]=played_last
matches["losers"]=losers
```

```python
winner_df=matches[matches["winner"]==matches["chasing"]]
print(winner_df["winner"].value_counts().idxmax(),winner_df["winner"].value_counts().max())
losers_df=matches[matches["losers"]==matches["chasing"]]
print(losers_df["losers"].value_counts().idxmax(),losers_df["losers"].value_counts().max())


```

Which toss decision is best for each ground?

```python

winners=matches[matches["toss_winner"]==matches["winner"]].groupby(["venue","toss_decision"]).count()['id']
dec_df=winners.unstack()
dec_df.fillna(0,inplace=True)
dec=[]
for i in range(len(dec_df)):
    if dec_df.iloc[i]["bat"]>dec_df.iloc[i]["field"]:
        dec.append("bat")
    else:
        dec.append("field")
dec_df["decision"]=dec
dec_df

#winners.agg(["winner",]).count()
```
